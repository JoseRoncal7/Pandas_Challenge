## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

male_count=purchase_data[purchase_data["Gender"] == "Male"]["SN"].nunique()
female_count=purchase_data[purchase_data["Gender"] == "Female"]["SN"].nunique()
other_count=purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]["SN"].nunique()


male_percentage = ((male_count/576)*100)
female_percentage = ((female_count/576)*100)
other_percentage = ((other_count)/576)*100

full_count=purchase_data["SN"].nunique()


gender_table = pd.DataFrame({"Gender": ["Male","Female","Other/Non-Disclosed"],"Total Count":[male_count,female_count,other_count],
                             "Percentage of Players":[male_percentage,female_percentage,other_percentage]})

gender_table = gender_table.set_index("Gender")

gender_table["Percentage of Players"] = gender_table["Percentage of Players"].map("{:,.2f}%".format)

gender_table
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python


male_purchase = purchase_data[purchase_data["Gender"] == "Male"]["Price"].count()

print(f"Male Purchase Count : {male_purchase}")

female_purchase = purchase_data[purchase_data["Gender"] == "Female"]["Price"].count()

print(f"Female Purchase Count: {female_purchase}")

total_purchases = purchase_data["Price"].count()

other_purchase = total_purchases - male_purchase - female_purchase

print(f"Other Purchase Count: {other_purchase}")

print("-------------------")

male_price_avg = purchase_data[purchase_data["Gender"] == "Male"]["Price"].mean()
male_price_avg = male_price_avg.round(2)

print(f"Male avg purchase price: {male_price_avg}")

female_price_avg = purchase_data[purchase_data["Gender"] == "Female"]["Price"].mean()
female_price_avg = female_price_avg.round(2)

print(f"Female avg purchase price: {female_price_avg}")

other_price_avg = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]["Price"].mean()
other_price_avg = other_price_avg.round(2)

print(f"Other avg purchase price: {other_price_avg}")

print("-------------------")

male_purchase_total = purchase_data[purchase_data["Gender"] == "Male"]["Price"].sum()
male_purchase_total = male_purchase_total.round(2)

print(f"Male Total purchase value: {male_purchase_total}")

female_purchase_total = purchase_data[purchase_data["Gender"] == "Female"]["Price"].sum()
female_purchase_total = female_purchase_total.round(2)

print(f"Female Total purchase value: {female_purchase_total}")

other_purchase_total = purchase_data[purchase_data["Gender"] == "Other / Non-Disclosed"]["Price"].sum()
other_purchase_total = other_purchase_total.round(2)

print(f"Other Total purchase value: {other_purchase_total}")

print("-------------------")

male_avg_total = male_purchase_total/male_count
male_avg_total=male_avg_total.round(2)
print(f"Male avg Total Purchase per person: {male_avg_total}")

female_avg_total = female_purchase_total/female_count
female_avg_total=female_avg_total.round(2)
print(f"Female avg Total Purchase per person: {female_avg_total}")

other_avg_total = other_purchase_total/other_count
other_avg_total=other_avg_total.round(2)
print(f"Other avg Total Purchase per person: {other_avg_total}")

gender_purchase_df = pd.DataFrame({"Gender": ["Male","Female","Other/Non-Disclosed"],"Purchase Count":[male_purchase,female_purchase
,other_purchase],"Average Purchase Price":[male_price_avg,female_price_avg,other_price_avg],"Total Purchase Value":[male_purchase_total,
female_purchase_total,other_purchase_total],"Avg Total Purchase per Person":[male_avg_total,female_avg_total,other_avg_total]})

gender_purchase_df = gender_purchase_df.set_index("Gender")

gender_purchase_df["Average Purchase Price"] = gender_purchase_df["Average Purchase Price"].map("${:,.2f}".format)

gender_purchase_df["Avg Total Purchase per Person"] = gender_purchase_df["Avg Total Purchase per Person"].map("${:,.2f}".format)

gender_purchase_df["Total Purchase Value"] = gender_purchase_df["Total Purchase Value"].map("${:,.2f}".format)

gender_purchase_df
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sn_total_purch = purchase_data.groupby('SN')['Price'].sum().to_frame()

sn_purch_count = purchase_data.groupby('SN')['Price'].count().to_frame()

sn_avg_purch = purchase_data.groupby('SN')['Price'].mean().to_frame()
sn_avg_purch = sn_avg_purch.round(2)

sn_total_purch.columns=["Total Purchase Value"]

first_add = sn_total_purch.join(sn_purch_count,how="left")
first_add.columns=["Total Purchase Value","Purchase Count"]

second_add = first_add.join(sn_avg_purch,how="inner")
second_add.columns=["Total Purchase Value","Purchase Count","Average Purchase Price"]

top_spenders_df = second_add[["Purchase Count", "Average Purchase Price", "Total Purchase Value"]]
top_spenders_df = top_spenders_df.sort_values('Total Purchase Value', ascending=False).head()

top_spenders_df["Average Purchase Price"] = top_spenders_df["Average Purchase Price"].map("${:,.2f}".format)
top_spenders_df["Total Purchase Value"] = top_spenders_df["Total Purchase Value"].map("${:,.2f}".format)
top_spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

premergeone = purchase_data.groupby("Item Name").sum().reset_index()

premergetwo = purchase_data.groupby("Item ID").sum().reset_index()
premergethree = purchase_data.groupby("Item Name").count().reset_index()


mergeone = pd.merge(premergeone, premergetwo, on="Price")
mergetwo = pd.merge(premergethree, mergeone, on="Item Name")


mergetwo["Gender"] = (mergetwo["Price_y"]/mergetwo["Item ID"]).round(2)

mergetwo_renamed = mergetwo.rename(columns={"Age": "Purchase Count", "Gender": "Item Price", "Item ID": "null", "Price_y": "Total Purchase Value", "Item ID_y": "Item ID"})


clean_df = mergetwo_renamed[["Item ID", "Item Name", "Purchase Count", "Item Price", "Total Purchase Value"]]

prefinal_df = clean_df.set_index(['Item Name', 'Item ID'])
popular_items_final = prefinal_df.sort_values('Purchase Count', ascending=False).head(6)
popular_items_final.style.format({"Item Price": "${:.2f}", "Total Purchase Value": "${:.2f}"})

```
